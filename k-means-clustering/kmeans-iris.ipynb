# K Means Clustering on Iris Dataset

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data)
```

```python
df
```

```python
iris.feature_names
```

As we will only be using Petal features (Width and Height), we drop Sepal Features

```python
df = df[2:]
df
```

```python
cluster = KMeans(n_clusters=3)
```

```python
cluster.fit(df)
```

```python
df['pseudo_labels']=cluster.predict(df)
```

```python
df['labels'] = pd.DataFrame(iris.target)
```

```python
plt.figure(dpi=150)
sns.scatterplot(data=df,x=2,y=3,hue='labels')
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Width classified by true label")
```

```python
plt.figure(dpi=150)
sns.scatterplot(data=df,x=2,y=3,hue='pseudo_labels')
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Width classified by K Means Clustering label")
```
