```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('spam.csv')
```

```python
df.head()
```

```python
df['Spam'] = pd.get_dummies(df['Category'],drop_first=True)
df
```

```python
df = df.drop('Category',axis=1)
```

```python
df
```

```python
x = df.drop('Spam',axis=1)
y=df['Spam']
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
xtrain
```

```python
cv = CountVectorizer()
```

```python
xtrain_count = cv.fit_transform(xtrain['Message'])
```

```python
xtrain_count.toarray()[:3]
```

```python
xtrain_count
```

```python
model = MultinomialNB()
model.fit(xtrain_count,ytrain)
```

```python
xtest_count = cv.transform(xtest['Message'])
mnbpred = model.predict(xtest_count)
accuracy_score(mnbpred,ytest)
```

Therefore Naive Bayes Spam Classifier is **98.39%** accuracte when predicting spam emails

Using **tf-idf Vectorizer**

```python
tfidf = TfidfVectorizer()
xtrain_tfidf = tfidf.fit_transform(xtrain['Message'])
xtest_tfidf  = tfidf.transform(xtest['Message'])
```

```python
nbc = MultinomialNB()
nbc.fit(xtrain_tfidf,ytrain)
tfidfpreds = nbc.predict(xtest_tfidf)
accuracy_score(tfidfpreds,ytest)
```

Therefore Naive Bayes Spam Classifier is **98.39%** accuracte when predicting spam emails using TFIDF Vectorizer

```python
wines = load_wine()
wines
```

```python
winedf = pd.DataFrame(wines.data,columns=wines.feature_names)
```

```python
winedf
```

```python
winedf['target'] = pd.DataFrame(wines.target)
winedf
```

```python
wines.target_names
```

```python
wxtrain,wxtest,wytrain,wytest = train_test_split(winedf.drop('target',axis=1),winedf['target'],test_size=0.2,random_state=101)
```

```python
winetester = MultinomialNB()
winetester.fit(wxtrain,wytrain)
winepred = winetester.predict(wxtest)
accuracy_score(winepred,wytest)
```

```python
winetester = GaussianNB()
winetester.fit(wxtrain,wytrain)
winepred = winetester.predict(wxtest)
accuracy_score(winepred,wytest)
```
